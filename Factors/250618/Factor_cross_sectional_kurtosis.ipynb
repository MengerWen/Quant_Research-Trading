```python
ds=DataService()
df=ds['ETHUSDT_15m_2020_2025']['2021-10-01':]
```

```python
evaluator=FactorEvaluation(df=df,future_return_periods=10)
```

```python
def cross_sectional_kurtosis_factor(df, window=20):
    returns = df['close'].pct_change()
    kurtosis = returns.rolling(window=window).kurt()
    cross_sectional_kurtosis = kurtosis - kurtosis.mean()
    return cross_sectional_kurtosis
```

```python
evaluator.set_factor(
    factor_data_or_func=cross_sectional_kurtosis_factor,
    factor_name='cross_sectional_kurtosis_factor'
)
```

```python
result=evaluator.run_full_evaluation(run_stationarity_test=True)
```
