```python
ds=DataService()
df=ds['ETHUSDT_15m_2020_2025']['2021-10-01':]
```

```python
evaluator=FactorEvaluation(df=df,future_return_periods=10)
```

```python
def ma_channel_breakout(df, ma_period=20, channel_width=0.05):
    ma = df['close'].rolling(ma_period).mean()
    upper_band = ma * (1 + channel_width)
    lower_band = ma * (1 - channel_width)
    
    # çªç ´ä¸Šè½¨ä¸ºæ­£ï¼Œçªç ´ä¸‹è½¨ä¸ºè´Ÿ
    above_upper = (df['close'] > upper_band).astype(int)
    below_lower = (df['close'] < lower_band).astype(int) * -1
    return above_upper + below_lower
```

```python
evaluator.set_factor(
    factor_data_or_func=ma_channel_breakout,
    factor_name='ma_channel_breakout'
)
```

```python
result=evaluator.run_full_evaluation(run_stationarity_test=True)
```
