# ÂÆö‰πâÂõ†Â≠êÔºÅ

```python
def factor(df):
    """
    Èáè‰ª∑ÂçèÂêåÊî∂ÊïõÂõ†Â≠ê (ËµÑÈáëÊµÅÂ¢ûÂº∫Áâà)
    ‰ºòÂåñÁÇπÔºö
    1. ÂºïÂÖ•‰∏ªÂä®‰π∞ÂÖ•ÈáèÂç†ÊØîÊåáÊ†á
    2. ÁªìÂêàËµÑÈáëÊµÅ‰∏éÈáè‰ª∑ËÉåÁ¶ª
    3. ‰ΩøÁî®ÊåáÊï∞Âä†ÊùÉÁßªÂä®Âπ≥Âùá
    """
    # ËÆ°ÁÆó‰∏ªÂä®‰π∞ÂÖ•ÈáèÂç†ÊØî
    df['buy_ratio'] = df['taker_buy_volume'] / (df['volume'] + 1e-7)
    
    # ËÆ°ÁÆó‰ª∑Ê†ºÊ≥¢Âä®Áéá
    price_change = df['close'].pct_change()
    volatility = price_change.ewm(span=96, min_periods=24).std()
    
    # ËÆ°ÁÆóÈáè‰ª∑ÊØîÁéá
    vp_ratio = df['volume'] / (volatility + 1e-7)
    
    # ÂÖ≥ÈîÆÂë®ÊúüÂÆö‰πâ
    periods = [96, 288, 672, 1440, 2880]
    
    # ËÆ°ÁÆóÂêÑÂë®ÊúüÈáè‰ª∑ÊØîÁéá
    vp_matrix = np.column_stack([
        vp_ratio.ewm(span=p, min_periods=int(p/4)).mean()  # ‰ΩøÁî®EMA
        for p in periods
    ])
    
    # ËÆ°ÁÆóÈáè‰ª∑-ËµÑÈáëÊµÅÂçèÂêåÊåáÊ†á
    synergy_matrix = np.zeros((len(df), len(periods)))
    for i, p in enumerate(periods):
        # ‰ª∑Ê†ºË∂ãÂäø
        price_trend = df['close'].pct_change(p)
        # ËµÑÈáëÊµÅË∂ãÂäø
        flow_trend = df['buy_ratio'].diff(p)
        # Èáè‰ª∑ËµÑÈáëÊµÅÂçèÂêå (ÊñπÂêë‰∏ÄËá¥Êó∂Â¢ûÂº∫)
        synergy_matrix[:, i] = np.sign(price_trend) * flow_trend
    
    # ÁªÑÂêàÁâπÂæÅÁü©Èòµ
    combined_matrix = np.concatenate((vp_matrix, synergy_matrix), axis=1)
    
    # ÊªöÂä®ÂàÜ‰ΩçÊï∞Ê†áÂáÜÂåñ
    scaled_std = np.zeros_like(combined_matrix)
    for i in range(combined_matrix.shape[1]):
        col = pd.Series(combined_matrix[:, i])
        # ÊªöÂä®ÂàÜ‰ΩçÊï∞ËÆ°ÁÆó
        rolling_q20 = col.rolling(window=2880, min_periods=720).quantile(0.20)
        rolling_q80 = col.rolling(window=2880, min_periods=720).quantile(0.80)
        scaled_std[:, i] = (col - rolling_q20) / (rolling_q80 - rolling_q20 + 1e-7)
    
    # ... [ÂâçÈù¢ÈÉ®ÂàÜ‰∏éÂéüÁ®ãÂ∫èÁõ∏ÂêåÔºåÁõ¥Âà∞ÁªÑÂêàÁâπÂæÅÁü©Èòµ] ...
    
    # === Ê†∏ÂøÉÊîπËøõ3: Â∏ÇÂú∫Áä∂ÊÄÅÊ£ÄÊµã ===
    # ËÆ°ÁÆóÂ∏ÇÂú∫Áä∂ÊÄÅÊåáÊ†á (‰∏ç‰ΩøÁî®Êú™Êù•Êï∞ÊçÆ)
    market_state = pd.Series(np.zeros(len(df)), index=df.index)
    
    # 1. Ê≥¢Âä®ÁéáÁä∂ÊÄÅ
    vol_state = (volatility > volatility.rolling(1440).quantile(0.75)).astype(int)
    
    # 2. ËµÑÈáëÊµÅÁä∂ÊÄÅ
    flow_state = (df['buy_ratio'] < df['buy_ratio'].rolling(1440).quantile(0.25)).astype(int)
    
    # 3. ‰ª∑Ê†ºË∂ãÂäøÁä∂ÊÄÅ (Áü≠ÊúüvsÈïøÊúü)
    short_trend = df['close'].pct_change(96)
    long_trend = df['close'].pct_change(672)
    trend_state = ((np.sign(short_trend) != np.sign(long_trend))).astype(int)
    
    # ÁªºÂêàÂ∏ÇÂú∫Áä∂ÊÄÅ (0-Ê≠£Â∏∏, 1-Â∞æÈÉ®)
    market_state = ((vol_state + flow_state + trend_state) >= 2).astype(int)
    
    # ËÆ°ÁÆóÊ≠£Â∏∏Â∏ÇÂú∫Âõ†Â≠ê
    normal_factor = np.mean(scaled_std, axis=1) * np.log1p(df['volume']) * df['buy_ratio']
    
    # ËÆ°ÁÆóÂ∞æÈÉ®Â∏ÇÂú∫‰∏ìÁî®Âõ†Â≠ê
    # ‰ΩøÁî®Áü≠Êúü‰ø°Âè∑ + ÂèçËΩ¨ÈÄªËæë
    tail_synergy = synergy_matrix[:, 0]  # ‰ΩøÁî®ÊúÄÁü≠Âë®Êúü(96)
    tail_volume = np.log1p(df['volume'].rolling(96).mean())  # Áü≠ÊúüÊàê‰∫§Èáè
    
    # Âú®Â∞æÈÉ®Âå∫Âüü‰ΩøÁî®ÂèçËΩ¨ÈÄªËæë
    price_change = df['close'].pct_change(96)
    reversal_signal = -np.sign(price_change) * tail_volume
    
    tail_factor = tail_synergy * reversal_signal * (1 - df['buy_ratio'])
    
    # Âõ†Â≠êËûçÂêà
    factor = np.where(market_state == 1, tail_factor, normal_factor)
    return factor
```

# ÊµãËØïÂõ†Â≠êË°®Áé∞

```python
evaluator.set_factor(
    factor_func=lambda df: factor(df),
    factor_name='factor'
)

result=evaluator.run_full_evaluation(run_stationarity_test=True)
```
