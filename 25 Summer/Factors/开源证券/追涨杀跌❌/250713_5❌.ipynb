# ÂÆö‰πâÂõ†Â≠êÔºÅ

```python
# %% [markdown]
# # ÊîπËøõÂõ†Â≠ê3ÔºöÂ∏ÇÂú∫Áä∂ÊÄÅËá™ÈÄÇÂ∫îÂõ†Â≠ê

# %%
def factor(df):
    """
    ÊîπËøõÁÇπÔºö
    1. Â§öÊó∂Èó¥Â∞∫Â∫¶ÂàÜÊûêÔºöÁªìÂêàÁü≠„ÄÅ‰∏≠„ÄÅÈïøÊúüÂä®Èáè
    2. Â∏ÇÂú∫Áä∂ÊÄÅËØÜÂà´ÔºöÈÄöËøáÊ≥¢Âä®ÁéáÂíåÊàê‰∫§ÈáèËØÜÂà´Â∏ÇÂú∫Áä∂ÊÄÅ
    3. Ëá™ÈÄÇÂ∫îÂõ†Â≠êËûçÂêàÔºö‰∏çÂêåÂ∏ÇÂú∫Áä∂ÊÄÅ‰ΩøÁî®‰∏çÂêåÂõ†Â≠êÁªÑÂêà
    """
    # ËÆ°ÁÆóÂ§öÂ∞∫Â∫¶Êî∂ÁõäÁéá
    ret_short = df['close'].pct_change(16)    # 4Â∞èÊó∂
    ret_medium = df['close'].pct_change(96)   # 24Â∞èÊó∂
    ret_long = df['close'].pct_change(1920)   # 20Â§©
    
    # ËÆ°ÁÆóÊ≥¢Âä®ÁéáÊåáÊ†á
    vol_short = df['close'].rolling(16).std()
    vol_medium = df['close'].rolling(96).std()
    
    # ËÆ°ÁÆóÊàê‰∫§ÈáèÊåáÊ†á
    vol_ma_short = df['volume'].rolling(16).mean()
    vol_ma_long = df['volume'].rolling(96).mean()
    volume_ratio = vol_ma_short / (vol_ma_long + 1e-8)
    
    # Â∏ÇÂú∫Áä∂ÊÄÅÂàÜÁ±ª
    market_state = pd.Series('Normal', index=df.index)
    
    # È´òÊ≥¢Âä®Áä∂ÊÄÅ (Ê≥¢Âä®Áéá > 2ÂÄç‰∏≠ÊúüÊ≥¢Âä®‰∏≠‰ΩçÊï∞)
    high_vol_threshold = vol_medium.median() * 2
    market_state[vol_short > high_vol_threshold] = 'HighVol'
    
    # È´òÊàê‰∫§ÈáèÁä∂ÊÄÅ
    market_state[volume_ratio > 2] = 'HighVolume'
    
    # Ë∂ãÂäøÁä∂ÊÄÅ (ÈïøÊúüÂä®ÈáèÊòæËëó)
    trend_threshold = vol_medium.median() * 1.5
    market_state[np.abs(ret_long) > trend_threshold] = 'Trending'
    
    # ÂàõÂª∫Áä∂ÊÄÅËá™ÈÄÇÂ∫îÂõ†Â≠ê
    factor = pd.Series(np.nan, index=df.index)
    
    # ‰∏çÂêåÁä∂ÊÄÅ‰ΩøÁî®‰∏çÂêåÂõ†Â≠êÁªÑÂêà
    for state in ['HighVol', 'HighVolume', 'Trending', 'Normal']:
        mask = (market_state == state)
        
        if state == 'HighVol':
            # Âú®È´òÊ≥¢Âä®Êúü‰ΩøÁî®ÂèçËΩ¨Á≠ñÁï•
            factor.loc[mask] = -ret_short.loc[mask]
            
        elif state == 'HighVolume':
            # Âú®È´òÊàê‰∫§ÈáèÊúü‰ΩøÁî®‰π∞ÂçñÂéãÂäõÊåáÊ†á
            buy_ratio = df['taker_buy_volume'] / (df['volume'] + 1e-8)
            factor.loc[mask] = ret_short.loc[mask] * buy_ratio.loc[mask]
            
        elif state == 'Trending':
            # Âú®Ë∂ãÂäøÊúü‰ΩøÁî®Â§öÂ∞∫Â∫¶Âä®ÈáèÁªÑÂêà
            factor.loc[mask] = (0.5 * ret_medium.loc[mask] + 
                               0.3 * ret_long.loc[mask] + 
                               0.2 * ret_short.loc[mask])
            
        else:  # Normal
            # Ê≠£Â∏∏Áä∂ÊÄÅ‰ΩøÁî®Ê≥¢Âä®ÁéáË∞ÉÊï¥Âä®Èáè
            adj_ret = ret_medium.loc[mask] / (vol_medium.loc[mask] + 1e-8)
            factor.loc[mask] = adj_ret
    
    # Âπ≥ÊªëÂõ†Â≠ê
    factor = factor.ewm(span=24).mean()
    
    return factor
```

# ÊµãËØïÂõ†Â≠êË°®Áé∞

```python
evaluator.set_factor(
    factor_func=lambda df: factor(df),
    factor_name='factor'
)

result=evaluator.run_full_evaluation(run_stationarity_test=True)
```
