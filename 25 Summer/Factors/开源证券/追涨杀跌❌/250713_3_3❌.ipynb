# å®šä¹‰å› å­ï¼

```python
# %% [markdown]
# # æ”¹è¿›å› å­4ï¼šå¤šå°ºåº¦èåˆè‡ªç›¸å…³å› å­

# %%
def factor(df):
    """
    æ”¹è¿›ç‚¹ï¼š
    1. å¤šæ—¶é—´å°ºåº¦åˆ†æï¼ˆé«˜é¢‘åˆ°ä½é¢‘ï¼‰
    2. é‡èƒ½ç¡®è®¤æœºåˆ¶
    3. ç›¸å…³æ€§è¡°å‡æ¨¡å‹
    """
    # å¤šå°ºåº¦æ”¶ç›Šç‡è®¡ç®—
    periods = [1, 4, 16, 64]  # 15m, 1h, 4h, 16h
    multi_scale_rets = {}
    for p in periods:
        multi_scale_rets[f'ret_{p}'] = np.log(df['close']).diff(p)
    
    # é‡èƒ½ç¡®è®¤æŒ‡æ ‡
    volume_conf = df['volume'] / df['volume'].rolling(96).mean()
    turnover_conf = df['turnover'] / df['turnover'].rolling(96).mean()
    conf_score = np.sqrt(volume_conf * turnover_conf)
    
    # å¤šå°ºåº¦è‡ªç›¸å…³è®¡ç®—
    autocorr_components = []
    for name, ret in multi_scale_rets.items():
        # é‡èƒ½è°ƒæ•´æ”¶ç›Šç‡
        adj_ret = ret * conf_score
        
        # è‡ªé€‚åº”çª—å£ï¼ˆä¸æ—¶é—´å°ºåº¦æˆæ­£æ¯”ï¼‰
        window = max(1920 // int(name.split('_')[1]), 96)  # ç¡®ä¿æœ€å°çª—å£
        
        # è¡°å‡åŠ æƒè‡ªç›¸å…³
        decay_factor = 0.99
        weights = np.array([decay_factor**i for i in range(window)][::-1])
        
        def decayed_autocorr(series):
            if len(series) < window:
                return np.nan
            lagged = series.shift(1).iloc[-window:]
            current = series.iloc[-window:]
            valid_mask = (~current.isna()) & (~lagged.isna())
            if valid_mask.sum() < 10:
                return np.nan
            cov = np.cov(current[valid_mask], lagged[valid_mask], aweights=weights[-len(current[valid_mask]):])[0, 1]
            var = np.var(current[valid_mask], ddof=0)
            return cov / var if var > 1e-8 else np.nan
        
        autocorr = adj_ret.rolling(window).apply(decayed_autocorr, raw=False)
        autocorr_components.append(autocorr)
    
    # å› å­èåˆï¼ˆåŠ æƒå¹³å‡ï¼‰
    weights = [0.4, 0.3, 0.2, 0.1]  # æ›´é«˜æƒé‡ç»™çŸ­æœŸå°ºåº¦
    combined = pd.DataFrame(autocorr_components).T
    factor = combined.dot(weights)
    
    # æµåŠ¨æ€§è¿‡æ»¤
    median_volume = df['volume'].rolling(2880).median()
    factor[df['volume'] < 0.5 * median_volume] = np.nan
    
    return factor
```

```python
evaluator.set_factor(
    factor_func=lambda df: factor(df),
    factor_name='factor'
)

result=evaluator.run_full_evaluation(run_stationarity_test=True)
```
