# ÂÆö‰πâÂõ†Â≠êÔºÅ

```python
def factor(df):
    """
    ËÆ°ÁÆóÂÖ®Êó∂ÊÆµËøΩÊ∂®ÊùÄË∑åÂõ†Â≠êÔºàÂü∫‰∫éË∂ÖÈ¢ùÊî∂ÁõäÂ∫èÂàóÁöÑ‰ΩôÂº¶Áõ∏‰ººÂ∫¶Ôºâ
    Ê≠•È™§Ôºö
    1. ËÆ°ÁÆó15ÂàÜÈíüÊî∂ÁõäÁéá
    2. ËÆ°ÁÆóÊªöÂä®ÂùáÂÄº‰Ωú‰∏∫Âü∫ÂáÜÊî∂ÁõäÁéáÔºàÊ®°ÊãüÂ∏ÇÂú∫Êî∂ÁõäÔºâ
    3. ËÆ°ÁÆóË∂ÖÈ¢ùÊî∂ÁõäÁéá = Êî∂ÁõäÁéá - Âü∫ÂáÜÊî∂ÁõäÁéá
    4. ËÆ°ÁÆóÂΩìÂâçÂë®ÊúüÂíå‰∏ã‰∏ÄÂë®ÊúüË∂ÖÈ¢ùÊî∂ÁõäÂ∫èÂàóÁöÑ‰ΩôÂº¶Áõ∏‰ººÂ∫¶
    """
    # ËÆ°ÁÆóÊî∂ÁõäÁéá
    ret = df['close'].pct_change()
    
    # ËÆ°ÁÆóÊªöÂä®ÂùáÂÄº‰Ωú‰∏∫Âü∫ÂáÜÊî∂ÁõäÁéáÔºà‰ΩøÁî®ËøáÂéª1920‰∏™Âë®Êúü‚âà20Â§©Ôºâ
    window_size = 1920  # 20Â§© * 96‰∏™15ÂàÜÈíüÂë®Êúü/Â§©
    benchmark_ret = ret.rolling(window=window_size, min_periods=1).mean()
    
    # ËÆ°ÁÆóË∂ÖÈ¢ùÊî∂ÁõäÁéá
    excess_ret = ret - benchmark_ret
    
    # ÊûÑÂª∫ÂΩìÂâçÂë®ÊúüÂíå‰∏ã‰∏ÄÂë®ÊúüÁöÑË∂ÖÈ¢ùÊî∂ÁõäÂ∫èÂàó
    current_er = excess_ret.shift(1)  # ÂΩìÂâçÂë®ÊúüË∂ÖÈ¢ùÊî∂Áõä
    next_er = excess_ret              # ‰∏ã‰∏ÄÂë®ÊúüË∂ÖÈ¢ùÊî∂Áõä
    
    # ËÆ°ÁÆó‰ΩôÂº¶Áõ∏‰ººÂ∫¶
    numerator = (current_er * next_er).rolling(window=window_size).sum()
    denom_current = (current_er ** 2).rolling(window=window_size).sum()
    denom_next = (next_er ** 2).rolling(window=window_size).sum()
    
    # ÈÅøÂÖçÈô§Èõ∂ÈîôËØØ
    denominator = np.sqrt(denom_current) * np.sqrt(denom_next)
    denominator = denominator.replace(0, np.nan)
    
    # ËÆ°ÁÆóÂõ†Â≠êÂÄº
    factor = numerator / denominator
    return factor
```

# ÊµãËØïÂõ†Â≠êË°®Áé∞

```python
evaluator.set_factor(
    factor_func=lambda df: factor(df),
    factor_name='factor'
)

result=evaluator.run_full_evaluation(run_stationarity_test=True)
```
