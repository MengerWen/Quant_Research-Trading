# ÂÆö‰πâÂõ†Â≠êÔºÅ

```python
def factor(df):
    """
    Èáè‰ª∑ÂçèÂêåÊî∂ÊïõÂõ†Â≠ê
    1. ÁªìÂêà‰ª∑Ê†ºÊ≥¢Âä®ÂíåÊàê‰∫§ÈáèÂèòÂåñ
    2. ËÆ°ÁÆóÈáè‰ª∑ËÉåÁ¶ªÊåáÊ†á
    3. ‰ΩøÁî®ÂàÜ‰ΩçÊï∞Ê†áÂáÜÂåñ
    """
    # ËÆ°ÁÆó‰ª∑Ê†ºÊ≥¢Âä®Áéá
    price_change = df['close'].pct_change().abs()
    volatility = price_change.rolling(window=96, min_periods=24).std()
    
    # ËÆ°ÁÆóÈáè‰ª∑ÊØîÁéá
    vp_ratio = df['volume'] / (volatility + 1e-7)
    
    # ÂÖ≥ÈîÆÂë®ÊúüÂÆö‰πâ
    periods = [96, 288, 672, 1440, 2880]  # 1d, 3d, 1w, 2w, 4w
    
    # ËÆ°ÁÆóÂêÑÂë®ÊúüÈáè‰ª∑ÊØîÁéá
    vp_matrix = np.column_stack([
        vp_ratio.rolling(window=p, min_periods=int(p/4)).mean() 
        for p in periods
    ])
    
    # ËÆ°ÁÆó‰ª∑Ê†ºÂíåÊàê‰∫§ÈáèÁöÑÁ¶ªÊï£Â∫¶ÊØîÁéá
    price_std = np.column_stack([
        df['close'].pct_change().rolling(window=p, min_periods=int(p/4)).std() 
        for p in periods
    ])
    
    volume_std = np.column_stack([
        df['volume'].pct_change().rolling(window=p, min_periods=int(p/4)).std() 
        for p in periods
    ])
    
    # ËÆ°ÁÆóÈáè‰ª∑Á¶ªÊï£Â∑ÆÂºÇ
    divergence = volume_std / (price_std + 1e-7)
    
    # ÁªÑÂêàÁâπÂæÅÁü©Èòµ
    combined_matrix = np.concatenate((vp_matrix, divergence), axis=1)
    
    # ËÆ°ÁÆóÂàÜ‰ΩçÊï∞Ê†áÂáÜÂåñÁ¶ªÊï£Â∫¶
    def quantile_scale(series):
        q25 = series.quantile(0.25)
        q75 = series.quantile(0.75)
        return (series - q25) / (q75 - q25 + 1e-7)
    
    scaled_std = np.apply_along_axis(
        lambda x: quantile_scale(pd.Series(x)).values, 
        0, 
        combined_matrix
    )
    
    # ÊúÄÁªàÂõ†Â≠êËÆ°ÁÆó
    factor = np.mean(scaled_std, axis=1) * np.log1p(df['volume'])
    return factor
```

# ÊµãËØïÂõ†Â≠êË°®Áé∞

```python
evaluator.set_factor(
    factor_func=lambda df: factor(df),
    factor_name='factor'
)

result=evaluator.run_full_evaluation(run_stationarity_test=True)
```
