# ÂÆö‰πâÂõ†Â≠êÔºÅ

```python
def factor(df):
    """
    Â§öÁ©∫ÂäõÈáèÊàê‰∫§ÈáèÊî∂ÊïõÂõ†Â≠ê
    1. Âå∫ÂàÜÂ§öÂ§¥ÂíåÁ©∫Â§¥Êàê‰∫§Èáè
    2. ËÆ°ÁÆóÂáÄÊµÅÈáèÊî∂ÊïõÂ∫¶
    3. Ê∑ªÂä†‰π∞ÂçñÂéãÂäõÊåáÊ†á
    """
    # ËÆ°ÁÆóÂáÄÊàê‰∫§Èáè (Â§öÂ§¥ - Á©∫Â§¥)
    buy_volume = df['taker_buy_volume']
    sell_volume = df['volume'] - buy_volume
    net_volume = buy_volume - sell_volume
    
    # ËÆ°ÁÆó‰π∞ÂçñÂéãÂäõÊØîÁéá
    pressure_ratio = (buy_volume / (sell_volume + 1e-7)).replace([np.inf, -np.inf], 0)
    
    # ÂÖ≥ÈîÆÂë®ÊúüÂÆö‰πâ
    periods = [96, 288, 672, 2016, 4032]  # 1d, 3d, 1w, 3w, 6w
    
    # ËÆ°ÁÆóÂáÄÊµÅÈáèÁöÑÂ§öÂë®ÊúüÊ†áÂáÜÂ∑Æ
    ma_values = [net_volume.rolling(window=p, min_periods=int(p/4)).mean() for p in periods]
    ma_matrix = np.column_stack(ma_values)
    
    # ËÆ°ÁÆóÂéãÂäõÊØîÁéáÁöÑÊ†áÂáÜÂ∑Æ
    pressure_ma = [pressure_ratio.rolling(window=p, min_periods=int(p/4)).mean() for p in periods]
    pressure_matrix = np.column_stack(pressure_ma)
    
    # ÁªÑÂêàÁâπÂæÅÁü©Èòµ
    combined_matrix = np.concatenate((ma_matrix, pressure_matrix), axis=1)
    
    # ËÆ°ÁÆóÁ¶ªÊï£Â∫¶ (‰ΩøÁî®MADÂ¢ûÂº∫È≤ÅÊ£íÊÄß)
    median = np.median(combined_matrix, axis=1)
    mad = np.median(np.abs(combined_matrix - median[:, np.newaxis]), axis=1)
    scaled_mad = mad / (np.abs(median) + 1e-7)
    
    # ÊñπÂêëÊÄßË∞ÉÊï¥ (ÁªìÂêàÂΩìÂâçÂáÄÊµÅÈáèÊñπÂêë)
    direction = np.sign(net_volume)
    factor = direction * np.log(1 + scaled_mad)
    return -abs(factor)
```

# ÊµãËØïÂõ†Â≠êË°®Áé∞

```python
evaluator.set_factor(
    factor_func=lambda df: factor(df),
    factor_name='factor'
)

result=evaluator.run_full_evaluation(run_stationarity_test=True)
```
