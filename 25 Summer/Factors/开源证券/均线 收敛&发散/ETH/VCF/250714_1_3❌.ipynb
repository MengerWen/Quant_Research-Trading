# ÂÆö‰πâÂõ†Â≠êÔºÅ

```python
def factor(df):
    """
    ‰ºòÂåñÁÇπÔºö
    1. ÂºïÂÖ•ËÆ¢ÂçïÊµÅ‰∏çÂπ≥Ë°°ÊåáÊ†á
    2. Â¢ûÂä†È´òÈ¢ëÁâπÂæÅÔºà‰π∞ÂçñÂéãÂäõÔºâ
    3. ÁªìÂêàÊàê‰∫§ÈáèÂàÜÂ∏ÉÁâπÂæÅ
    """
    # ÊîπËøõÁöÑ‰∏ªÂä®‰π∞ÂÖ•ÈáèÂç†ÊØîÔºàËÄÉËôëËÆ¢ÂçïÊµÅ‰∏çÂπ≥Ë°°Ôºâ
    df['taker_sell_volume'] = df['volume'] - df['taker_buy_volume']
    df['order_flow_imbalance'] = (df['taker_buy_volume'] - df['taker_sell_volume']) / (df['volume'] + 1e-7)
    df['buy_ratio'] = df['order_flow_imbalance'].ewm(span=24).mean()
    
    # ËÆ°ÁÆó‰π∞ÂçñÂéãÂäõÊåáÊ†áÔºàÈ´òÈ¢ëÁâπÂæÅÔºâ
    df['price_pressure'] = (2*df['close'] - df['open'] - df['low']) / (df['high'] - df['low'] + 1e-7)
    df['buy_pressure'] = df['price_pressure'] * df['taker_buy_volume']
    df['sell_pressure'] = (1 - df['price_pressure']) * df['taker_sell_volume']
    
    # ÂáÄÂéãÂäõÊåáÊ†á
    df['net_pressure'] = (df['buy_pressure'] - df['sell_pressure']).ewm(span=48).mean()
    
    # Ê≥¢Âä®ÁéáËÆ°ÁÆóÔºàÊï¥ÂêàÂéãÂäõÊåáÊ†áÔºâ
    range_vol = (df['high'] - df['low']).ewm(span=96).mean()
    pressure_vol = df['net_pressure'].abs().ewm(span=96).mean()
    df['volatility'] = np.sqrt(range_vol**2 + pressure_vol**2)
    
    # Êàê‰∫§ÈáèÂàÜÂ∏ÉÁâπÂæÅ
    df['volume_z'] = (df['volume'] - df['volume'].rolling(2880).mean()) / (df['volume'].rolling(2880).std() + 1e-7)
    volume_regime = np.select(
        [df['volume_z'] > 1.5, df['volume_z'] < -1.5, np.abs(df['volume_z']) <= 0.5],
        [2, 0.5, 1],  # È´ò/‰Ωé/Ê≠£Â∏∏Êàê‰∫§ÈáèÁ≥ªÊï∞
        default=1
    )
    
    # ËÆ°ÁÆóÈáè‰ª∑ÊØîÁéáÔºàËÄÉËôëÊàê‰∫§ÈáèÂàÜÂ∏ÉÔºâ
    vp_ratio = df['volume'] / (df['volatility'] + 1e-7) * volume_regime
    
    # ÂÖ≥ÈîÆÂë®ÊúüÂÆö‰πâ
    periods = [96, 288, 672, 1440, 2880]
    
    # ËÆ°ÁÆóÂêÑÂë®ÊúüÈáè‰ª∑ÊØîÁéá
    vp_matrix = np.column_stack([
        vp_ratio.ewm(span=p, min_periods=int(p/4)).mean()
        for p in periods
    ])
    
    # ÊîπËøõÁöÑÂçèÂêåÊåáÊ†áÔºàÊï¥ÂêàÂéãÂäõÊåáÊ†áÔºâ
    synergy_matrix = np.zeros((len(df), len(periods)))
    for i, p in enumerate(periods):
        # ‰ª∑Ê†ºË∂ãÂäøÔºàÁªìÂêàÂéãÂäõÊåáÊ†áÔºâ
        price_trend = df['close'].pct_change(p) + 0.2 * df['net_pressure'].diff(p)
        
        # ËµÑÈáëÊµÅË∂ãÂäøÔºà‰ΩøÁî®ËÆ¢ÂçïÊµÅ‰∏çÂπ≥Ë°°Ôºâ
        flow_trend = df['order_flow_imbalance'].diff(p)
        
        # Èáè‰ª∑ËµÑÈáëÊµÅÂçèÂêå
        synergy_matrix[:, i] = np.sign(price_trend) * flow_trend * volume_regime
    
    # ÁªÑÂêàÁâπÂæÅÁü©Èòµ
    combined_matrix = np.concatenate((vp_matrix, synergy_matrix), axis=1)
    
    # ÊîπËøõÁöÑÊ†áÂáÜÂåñÔºàÂàÜ‰ΩçÊï∞+Z-scoreÊ∑∑ÂêàÔºâ
    scaled_std = np.zeros_like(combined_matrix)
    for i in range(combined_matrix.shape[1]):
        col = pd.Series(combined_matrix[:, i])
        
        # ÂàÜ‰ΩçÊï∞Ê†áÂáÜÂåñ
        rolling_q25 = col.rolling(window=1440, min_periods=360).quantile(0.25)
        rolling_q75 = col.rolling(window=1440, min_periods=360).quantile(0.75)
        iqr = rolling_q75 - rolling_q25
        
        # Z-scoreÊ†áÂáÜÂåñ
        rolling_mean = col.rolling(window=1440, min_periods=360).mean()
        rolling_std = col.rolling(window=1440, min_periods=360).std()
        
        # Ê∑∑ÂêàÊ†áÂáÜÂåñÔºàIQRÁ®≥ÂÆöÔºåZ-scoreÊïèÊÑüÔºâ
        scaled_col = np.where(
            iqr > 0.1 * rolling_std, 
            (col - rolling_q25) / (iqr + 1e-7),  # ‰ΩøÁî®ÂàÜ‰ΩçÊï∞
            (col - rolling_mean) / (rolling_std + 1e-7)  # ‰ΩøÁî®Z-score
        )
        scaled_std[:, i] = scaled_col
    
    # ÊúÄÁªàÂõ†Â≠êËÆ°ÁÆóÔºàÂéãÂäõÂä†ÊùÉÔºâ
    pressure_weight = 1 + np.tanh(df['net_pressure'] * 2)  # ÂéãÂäõÊñπÂêëÂä†Âº∫‰ø°Âè∑
    flow_weight = df['buy_ratio'].clip(0.3, 0.7)
    
    factor = np.mean(scaled_std, axis=1) * np.log1p(df['volume']) * flow_weight * pressure_weight
    return factor
```

# ÊµãËØïÂõ†Â≠êË°®Áé∞

```python
evaluator.set_factor(
    factor_func=lambda df: factor(df),
    factor_name='factor'
)

result=evaluator.run_full_evaluation(run_stationarity_test=True)
```
