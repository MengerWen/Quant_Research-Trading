# ÂÆö‰πâÂõ†Â≠êÔºÅ

```python
def factor(df):
    """
    Èáè‰ª∑ÂçèÂêåÊî∂ÊïõÂõ†Â≠ê (Âä®ÊÄÅÂä†ÊùÉÁâà)
    ‰ºòÂåñÁÇπÔºö
    1. Ê†πÊçÆÂ∏ÇÂú∫Ê≥¢Âä®ÁéáÂä®ÊÄÅË∞ÉÊï¥Âë®ÊúüÊùÉÈáç
    2. Â§öÂ∞∫Â∫¶ÁâπÂæÅËûçÂêà
    3. Â¢ûÂä†Ë∂ãÂäøËøáÊª§Êú∫Âà∂
    """
    # ËÆ°ÁÆóÂ∏ÇÂú∫Ê≥¢Âä®Áä∂ÊÄÅ
    vol_short = df['close'].pct_change().rolling(96).std()
    vol_long = df['close'].pct_change().rolling(2880).std()
    vol_ratio = vol_short / (vol_long + 1e-7)
    
    # ËÆ°ÁÆóÂä®ÊÄÅÊùÉÈáç (È´òÊ≥¢Âä®Êó∂‰æßÈáçÁü≠Êúü‰ø°Âè∑)
    short_term_weight = np.tanh(vol_ratio * 2)  # 0-1‰πãÈó¥
    long_term_weight = 1 - short_term_weight
    
    # ËÆ°ÁÆó‰ª∑Ê†ºÊ≥¢Âä®Áéá
    price_change = df['close'].pct_change()
    volatility = price_change.rolling(window=96, min_periods=24).std()
    
    # ËÆ°ÁÆóÈáè‰ª∑ÊØîÁéá
    vp_ratio = df['volume'] / (volatility + 1e-7)
    
    # ÂÖ≥ÈîÆÂë®ÊúüÂÆö‰πâ
    short_periods = [96, 192, 288]  # Áü≠ÊúüÂë®Êúü
    long_periods = [672, 1440, 2880]  # ÈïøÊúüÂë®Êúü
    
    # ËÆ°ÁÆóÁü≠ÊúüÈáè‰ª∑ÊØîÁéá
    vp_short = np.column_stack([
        vp_ratio.rolling(window=p, min_periods=int(p/4)).mean() 
        for p in short_periods
    ])
    
    # ËÆ°ÁÆóÈïøÊúüÈáè‰ª∑ÊØîÁéá
    vp_long = np.column_stack([
        vp_ratio.rolling(window=p, min_periods=int(p/4)).mean() 
        for p in long_periods
    ])
    
    # ËÆ°ÁÆóÈáè‰ª∑ËÉåÁ¶ªÊåáÊ†á
    divergence = np.zeros((len(df), len(short_periods + long_periods)))
    all_periods = short_periods + long_periods
    for i, p in enumerate(all_periods):
        # ‰ª∑Ê†ºÂèòÂä®
        price_change = df['close'].diff(p)
        # Êàê‰∫§ÈáèÂèòÂä®
        vol_change = df['volume'].diff(p)
        # Èáè‰ª∑ËÉåÁ¶ª
        divergence[:, i] = np.abs(price_change) / (np.abs(vol_change) + 1e-7)
    
    # ÁªÑÂêàÁâπÂæÅÁü©Èòµ
    combined_matrix = np.concatenate((vp_short, vp_long, divergence), axis=1)
    
    # ÊªöÂä®ÂàÜ‰ΩçÊï∞Ê†áÂáÜÂåñ
    scaled_std = np.zeros_like(combined_matrix)
    for i in range(combined_matrix.shape[1]):
        col = pd.Series(combined_matrix[:, i])
        rolling_q25 = col.rolling(2880, min_periods=720).quantile(0.25)
        rolling_q75 = col.rolling(2880, min_periods=720).quantile(0.75)
        scaled_std[:, i] = (col - rolling_q25) / (rolling_q75 - rolling_q25 + 1e-7)
    
    # Â§öÂ∞∫Â∫¶ÁâπÂæÅËûçÂêà (Âä®ÊÄÅÂä†ÊùÉ)
    n_short = len(short_periods)
    n_long = len(long_periods)
    
    # Áü≠ÊúüÁâπÂæÅÈÉ®ÂàÜ
    short_features = scaled_std[:, :n_short]
    # ÈïøÊúüÁâπÂæÅÈÉ®ÂàÜ
    long_features = scaled_std[:, n_short:n_short+n_long]
    # ËÉåÁ¶ªÁâπÂæÅÈÉ®ÂàÜ
    divergence_features = scaled_std[:, n_short+n_long:]
    
    # Âä®ÊÄÅÂä†ÊùÉËûçÂêà
    short_component = np.mean(short_features, axis=1) * short_term_weight
    long_component = np.mean(long_features, axis=1) * long_term_weight
    divergence_component = np.mean(divergence_features, axis=1)
    
    # ÊúÄÁªàÂõ†Â≠êËÆ°ÁÆó
    factor = (0.4 * short_component + 
              0.4 * long_component + 
              0.2 * divergence_component) * np.log1p(df['volume'])
    
    # Ë∂ãÂäøËøáÊª§
    trend = df['close'].rolling(288).mean()
    factor = factor * np.sign(df['close'] - trend)  # ‰øùÁïôÊñπÂêë‰ø°ÊÅØ
    
    return factor
```

# ÊµãËØïÂõ†Â≠êË°®Áé∞

```python
evaluator.set_factor(
    factor_func=lambda df: factor(df),
    factor_name='factor'
)

result=evaluator.run_full_evaluation(run_stationarity_test=True)
```
