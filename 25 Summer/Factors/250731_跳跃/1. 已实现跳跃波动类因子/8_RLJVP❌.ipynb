# ÂÆö‰πâÂõ†Â≠êÔºÅ

```python
def factor(df):
    """
    ËÆ°ÁÆóRLJVPÂõ†Â≠ê (Â∑≤ÂÆûÁé∞‰∏äË°åÂ§ßÁ®ãË∑≥Ë∑ÉÊ≥¢Âä®)
    ÂÖ¨ÂºèÔºöRLJVP_t = min(RJVP_t, Œ£r_i^2 I_{r_i>Œ≥})
    """
    # ËÆ°ÁÆóÂØπÊï∞Êî∂ÁõäÁéá
    close_prices = df['close'].values
    log_returns = np.log(close_prices[1:]) - np.log(close_prices[:-1])
    
    # Â∏∏Êï∞
    mu1 = math.sqrt(2/math.pi)
    mu1_inv_sq = 1 / (mu1 ** 2)
    alpha = 4
    
    # ÂàùÂßãÂåñÁªìÊûúÊï∞ÁªÑ
    rljvp_values = np.full(len(df), np.nan)
    
    # ÊªöÂä®Á™óÂè£ËÆ°ÁÆó
    window_size = 96
    for i in range(window_size, len(df)):
        start_idx = i - window_size + 1
        end_idx = i
        
        window_returns = log_returns[start_idx:end_idx]
        n = len(window_returns)
        
        # ËÆ°ÁÆóBV
        abs_returns = np.abs(window_returns)
        bv = mu1_inv_sq * (n/(n-1)) * np.sum(abs_returns[:-1] * abs_returns[1:])
        
        # ËÆ°ÁÆóÊ≠£ÂêëÊ≥¢Âä®
        positive_returns = window_returns[window_returns > 0]
        rv_plus = np.sum(positive_returns ** 2)
        
        # ËÆ°ÁÆóRJVP
        rjvp = max(rv_plus - bv/2, 0)
        
        # ËÆ°ÁÆóÈòàÂÄºŒ≥
        gamma = alpha * (n ** (-0.49)) * math.sqrt(bv)
        
        # ËÆ°ÁÆóÂ§ßÁ®ãË∑≥Ë∑ÉÈÉ®ÂàÜ
        large_jumps = window_returns[window_returns > gamma]
        large_jump_variance = np.sum(large_jumps ** 2)
        
        # ËÆ°ÁÆóRLJVP
        rljvp = min(rjvp, large_jump_variance)
        rljvp_values[i] = rljvp
    
    return pd.Series(rljvp_values, index=df.index)
```

# ÊµãËØïÂõ†Â≠êË°®Áé∞

```python
evaluator.set_factor(
    factor_func=lambda df: factor(df),
    factor_name='factor'
)

result=evaluator.run_full_evaluation(run_stationarity_test=True)
```
