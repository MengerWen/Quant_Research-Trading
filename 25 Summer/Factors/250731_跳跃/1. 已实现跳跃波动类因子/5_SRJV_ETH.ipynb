```python
def factor(df):
    """
    è®¡ç®—SRJVå› å­ (ä¸Šä¸‹è¡Œè·³è·ƒæ³¢åŠ¨ä¸å¯¹ç§°)
    å…¬å¼ï¼šSRJV_t = RJVP_t - RJVN_t
    RJVP_t: ä¸Šè¡Œè·³è·ƒæ³¢åŠ¨
    RJVN_t: ä¸‹è¡Œè·³è·ƒæ³¢åŠ¨
    """
    # è®¡ç®—å¯¹æ•°æ”¶ç›Šç‡
    close_prices = df['close'].values
    log_returns = np.log(close_prices[1:]) - np.log(close_prices[:-1])
    
    # å¸¸æ•°
    mu1 = math.sqrt(2/math.pi)
    mu1_inv_sq = 1 / (mu1 ** 2)
    
    # åˆå§‹åŒ–ç»“æœæ•°ç»„
    srjv_values = np.full(len(df), np.nan)
    
    # æ»šåŠ¨çª—å£è®¡ç®—
    window_size = 96
    for i in range(window_size, len(df)):
        start_idx = i - window_size + 1
        end_idx = i
        
        window_returns = log_returns[start_idx:end_idx]
        n = len(window_returns)
        
        # è®¡ç®—BV
        abs_returns = np.abs(window_returns)
        bv = mu1_inv_sq * (n/(n-1)) * np.sum(abs_returns[:-1] * abs_returns[1:])
        
        # è®¡ç®—æ­£å‘æ³¢åŠ¨
        rv_plus = np.sum(window_returns[window_returns > 0] ** 2)
        
        # è®¡ç®—è´Ÿå‘æ³¢åŠ¨
        rv_minus = np.sum(window_returns[window_returns < 0] ** 2)
        
        # è®¡ç®—RJVPå’ŒRJVN
        rjvp = max(rv_plus - bv/2, 0)
        rjvn = max(rv_minus - bv/2, 0)
        
        # è®¡ç®—SRJV
        srjv = rjvp - rjvn
        srjv_values[i] = srjv
    
    return pd.Series(srjv_values, index=df.index)
```

# æµ‹è¯•å› å­è¡¨ç°

```python
evaluator.set_factor(
    factor_func=lambda df: factor(df),
    factor_name='factor'
)

result=evaluator.run_full_evaluation(run_stationarity_test=True)
```
