# ÂÆö‰πâÂõ†Â≠êÔºÅ

```python
def factor(df):
    """
    ËÆ°ÁÆóJarr_BNSÂõ†Â≠ê (Âü∫‰∫éBNSÊ£ÄÈ™åÁöÑË∑≥Ë∑ÉÂà∞ËææÁéá)
    ÂÖ¨ÂºèÔºöJarr = D^{-1} Œ£ I_{BNS,t}
    D: ËßÇÂØüÊúüÂ§©Êï∞ (30Â§©)
    I_{BNS,t}: BNSË∑≥Ë∑ÉÊ£ÄÈ™åÊåáÁ§∫ÂèòÈáè
    """
    # ËÆ°ÁÆóÂØπÊï∞Êî∂ÁõäÁéá
    close_prices = df['close'].values
    log_returns = np.log(close_prices[1:]) - np.log(close_prices[:-1])
    
    # Â∏∏Êï∞
    mu1 = math.sqrt(2/math.pi)
    mu1_inv_sq = 1 / (mu1 ** 2)
    mu_4_3 = (2**(2/3)) * gamma(7/6) / gamma(0.5)  # Œº_{4/3}
    
    # ÂàùÂßãÂåñÁªìÊûúÊï∞ÁªÑ
    jarr_bns_values = np.full(len(df), np.nan)
    
    # ËßÇÂØüÊúüÈïøÂ∫¶ (30Â§© = 30*96=2880Ê†πKÁ∫ø)
    observation_period = 2880
    
    for i in range(observation_period, len(df)):
        # Ëé∑ÂèñËßÇÂØüÊúüÁ™óÂè£ÂÜÖÁöÑÊï∞ÊçÆ
        start_idx = i - observation_period + 1
        end_idx = i
        window_returns = log_returns[start_idx:end_idx]
        
        # Â∞ÜÁ™óÂè£ÂàíÂàÜ‰∏∫30‰∏™‰∫§ÊòìÊó• (ÊØè‰∏™‰∫§ÊòìÊó•96Ê†πKÁ∫ø)
        daily_jumps = []
        
        for day_start in range(0, len(window_returns), 96):
            day_end = min(day_start + 96, len(window_returns))
            if day_end - day_start < 10:  # Á°Æ‰øùÊúâË∂≥Â§üÊï∞ÊçÆ
                continue
                
            daily_returns = window_returns[day_start:day_end]
            n = len(daily_returns)
            
            # ËÆ°ÁÆóRV (Â∑≤ÂÆûÁé∞Ê≥¢Âä®Áéá)
            rv = np.sum(daily_returns ** 2)
            
            # ËÆ°ÁÆóBV (ÂèåÂπÇÊ¨°ÂèòÂ∑Æ)
            abs_returns = np.abs(daily_returns)
            bv = mu1_inv_sq * (n/(n-1)) * np.sum(abs_returns[:-1] * abs_returns[1:])
            
            # ËÆ°ÁÆóTP (‰∏âÂπÇÊ¨°ÂèòÂ∑Æ)
            tp = 0
            for j in range(2, n):
                term = np.abs(daily_returns[j-2])**(4/3) * np.abs(daily_returns[j-1])**(4/3) * np.abs(daily_returns[j])**(4/3)
                tp += term
            tp = (mu_4_3**(-3)) * (n**2/(n-2)) * tp
            
            # ËÆ°ÁÆóT_BNSÁªüËÆ°Èáè
            denominator = ((math.pi/2)**2 + math.pi - 5) * (1/n) * max(1, tp/(bv**2))
            t_bns = (1 - bv/rv) / math.sqrt(denominator) if denominator > 0 else 0
            
            # Âà§Êñ≠ÊòØÂê¶Ë∑≥Ë∑É (99%ÁΩÆ‰ø°Ê∞¥Âπ≥)
            jump_indicator = 1 if t_bns > 2.33 else 0
            daily_jumps.append(jump_indicator)
        
        # ËÆ°ÁÆóË∑≥Ë∑ÉÂà∞ËææÁéá
        if daily_jumps:
            jarr = np.mean(daily_jumps)
            jarr_bns_values[i] = jarr
    
    return pd.Series(jarr_bns_values, index=df.index)
```

# ÊµãËØïÂõ†Â≠êË°®Áé∞

```python
evaluator.set_factor(
    factor_func=lambda df: factor(df),
    factor_name='factor'
)

result=evaluator.run_full_evaluation(run_stationarity_test=True)
```
