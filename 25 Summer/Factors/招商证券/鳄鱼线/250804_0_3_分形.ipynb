```python
def factor(df):
    """
    è®¡ç®—é³„é±¼çº¿ + AO + åˆ†å½¢ç»„åˆå› å­ï¼ˆè¿žç»­ç‰ˆæœ¬ï¼‰
    å› å­å€¼ = é³„é±¼çº¿å¼ºåº¦ * (AOåŠ¨é‡ + åˆ†å½¢çªç ´å¼ºåº¦)
    å…¶ä¸­ï¼š
    - é³„é±¼çº¿å¼ºåº¦ = (å”‡å»çº¿ - ä¸‹é¢šçº¿)
    - AOåŠ¨é‡ = AOçš„3æœŸå˜åŒ–
    - åˆ†å½¢çªç ´å¼ºåº¦ = (æ”¶ç›˜ä»· - æœ€è¿‘ä¸Šåˆ†å½¢æœ€é«˜ä»·) - (æœ€è¿‘ä¸‹åˆ†å½¢æœ€ä½Žä»· - æ”¶ç›˜ä»·)
    """
    # è®¡ç®—ä¸­é—´ä»·
    median_price = (df['high'] + df['low']) / 2
    
    # è®¡ç®—é³„é±¼çº¿
    lip = median_price.rolling(window=5).mean().shift(3)
    teeth = median_price.rolling(window=8).mean().shift(5)
    jaw = median_price.rolling(window=13).mean().shift(8)
    
    # è®¡ç®—é³„é±¼çº¿å¼ºåº¦
    alligator_strength = lip - jaw
    
    # è®¡ç®—AOæŒ‡æ ‡
    sma5 = median_price.rolling(window=5).mean()
    sma34 = median_price.rolling(window=34).mean()
    ao = sma5 - sma34
    
    # è®¡ç®—AOåŠ¨é‡
    ao_momentum = ao.diff(3)
    
    # è®¡ç®—åˆ†å½¢ä¿¡å·
    high = df['high']
    low = df['low']
    close = df['close']
    
    # è¯†åˆ«ä¸Šåˆ†å½¢
    up_fractal = (
        (high.shift(2) > high.shift(4)) & 
        (high.shift(2) > high.shift(3)) & 
        (high.shift(2) > high.shift(1)) & 
        (high.shift(2) > high))
    
    # è¯†åˆ«ä¸‹åˆ†å½¢
    down_fractal = (
        (low.shift(2) < low.shift(4)) & 
        (low.shift(2) < low.shift(3)) & 
        (low.shift(2) < low.shift(1)) & 
        (low.shift(2) < low))
    
    # èŽ·å–æœ€è¿‘çš„ä¸Šåˆ†å½¢æœ€é«˜ä»·
    last_up_fractal_high = high.copy()
    last_up_fractal_high[~up_fractal] = np.nan
    last_up_fractal_high = last_up_fractal_high.ffill()
    
    # èŽ·å–æœ€è¿‘çš„ä¸‹åˆ†å½¢æœ€ä½Žä»·
    last_down_fractal_low = low.copy()
    last_down_fractal_low[~down_fractal] = np.nan
    last_down_fractal_low = last_down_fractal_low.ffill()
    
    # è®¡ç®—åˆ†å½¢çªç ´å¼ºåº¦
    fractal_strength = (close - last_up_fractal_high) - (last_down_fractal_low - close)
    
    # ç»„åˆå› å­å€¼
    factor_value = alligator_strength * (ao_momentum + fractal_strength)
    
    return factor_value
```

# æµ‹è¯•å› å­è¡¨çŽ°

```python
evaluator.set_factor(
    factor_func=lambda df: factor(df),
    factor_name='factor'
)

result=evaluator.run_full_evaluation(run_stationarity_test=True)
```
